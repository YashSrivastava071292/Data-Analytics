Q1: Fisher Discriminant Analysis

```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading dataset
df_train = pd.read_csv("train.csv")
```

```python
#Extracting unique labels into list
lab_list = df_train['label'].unique()
str(lab_list)
```

```python
#Iterating the list to create pairs
res = [(a, b) for idx, a in enumerate(lab_list) for b in lab_list[idx + 1:]]
res
```

```python
len(res)
```

```python
res[0][0]
```

```python
#Looping through each pair and calculate fisher score
fisher_value = []
for i in range(len(res)):
    df = df_train[(df_train['label'] == res[i][0]) | (df_train['label'] == res[i][1])]
    c1 = df[df['label']==res[i][0]].count()
    c2 = df[df['label']==res[i][1]].count()
    df2_means = df.groupby(['label']).mean()
    df3_std = df.groupby(['label']).std()
    df_md =np.square(df2_means.iloc[1].subtract(df2_means.iloc[0]))#Between class seperation given by difference of squared means
    df_sd =c2*np.square(df3_std.iloc[1]) + c1*np.square(df3_std.iloc[0])#Within class standard deviation
    result= df_md/df_sd
    fisher_value.append(result)
```

```python
len(fisher_value)
```

```python
fisher_value
```

```python
#Converting into dataframe
df_new = pd.DataFrame(fisher_value)
df_new
```
