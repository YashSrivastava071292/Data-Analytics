```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.neighbors import KernelDensity
from sklearn.svm import SVC
```

```python
df = pd.read_csv('letter-recognition.csv')
df.head()
```

```python
unique_letter = list(np.sort(df['letter'].unique()))
print(unique_letter)
unique_letter_array = np.array(unique_letter)
```

```python
#Iterating the list to create pairs
res = [(a, b) for idx, a in enumerate(unique_letter_array) for b in unique_letter_array[idx + 1:]]
res
```

```python
#Looping through each pair and calculate fisher score
result = []
for i in range(len(res)):
    df2 = df[(df['letter'] == res[i][0]) | (df['letter'] == res[i][1])]
    X = df2.drop("letter", axis = 1)
    y = df2['letter']
    s = MinMaxScaler()
    #fit_transform for dataset
    X_scaled = s.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3, random_state = 101, stratify=y)
    model = SVC(C=5, kernel='linear')
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = metrics.accuracy_score(y_true=y_test, y_pred=y_pred)
    results = {'Positive class':res[i][0], 'Negative class':res[i][1], 'Linear(C=5)':accuracy}
    result.append(results)
    
```

```python
result
```

```python
df_new = pd.DataFrame(result)
df_new
```

```python
df_new.sort_values(by='Linear(C=5)', ascending=[True]).head()
```

```python
df_new.sort_values(by='Linear(C=5)', ascending=[False]).head()
```

As seen from above, when the model is sorted in ascending order the accuracy is lower hence the model is not able to differentiate. But for descending order it is able to differntiate.
