```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.neighbors import KernelDensity
from sklearn.svm import SVC
```

```python
df = pd.read_csv('letter-recognition.csv')
df.head()
```

```python
letter_uni = list(np.sort(df['letter'].unique()))
print(letter_uni)
```

```python
letters = df[['letter']]
X_train, X_test, y_train, y_test = train_test_split(df, letters, test_size=0.30, random_state=42,
                                                    stratify=letters)
pd.concat([df['letter'].value_counts(),y_train['letter'].value_counts(),y_test['letter'].value_counts()],
          axis = 1, sort=True).head(10)
```

```python
X_train = X_train.drop(['letter'], axis=1)
X_test  = X_test.drop(['letter'], axis=1)
```

```python
s = MinMaxScaler()
X_train  = s.fit_transform(X_train)
X_test  = s.fit_transform(X_test)
```

```python
result = []
for i in range(len(letter_uni)):
    y_train['Label'] = np.where(y_train['letter'] == letter_uni[i], '1', '0')
    y_test['Label'] = np.where(y_test['letter'] == letter_uni[i], '1', '0')
    y_train_2 = y_train.drop(['letter'], axis=1)
    y_test_2  = y_test.drop(['letter'], axis=1)
    
    model_linear1 = SVC(C=1, kernel='linear')
    model_linear1.fit(X_train, y_train_2)
    y_pred = model_linear1.predict(X_test)
    accuracy1 = metrics.accuracy_score(y_true=y_test_2, y_pred=y_pred)
    
    model_linear2 = SVC(C=5, kernel='linear')
    model_linear2.fit(X_train, y_train_2)
    y_pred = model_linear2.predict(X_test)
    accuracy2 = metrics.accuracy_score(y_true=y_test_2, y_pred=y_pred)
    
    model_linear3 = SVC(C=10, kernel='linear')
    model_linear3.fit(X_train, y_train_2)
    y_pred = model_linear3.predict(X_test)
    accuracy3 = metrics.accuracy_score(y_true=y_test_2, y_pred=y_pred)
    
    model_poly1 = SVC(C=5, kernel='poly', degree=2)
    model_poly1.fit(X_train, y_train_2)
    y_pred = model_poly1.predict(X_test)
    accuracy4 = metrics.accuracy_score(y_true=y_test_2, y_pred=y_pred)
    
    model_poly2 = SVC(C=5, kernel='poly', degree=3)
    model_poly2.fit(X_train, y_train_2)
    y_pred = model_poly2.predict(X_test)
    accuracy5 = metrics.accuracy_score(y_true=y_test_2, y_pred=y_pred)
    
    model_rbf1 = SVC(C=5, kernel='rbf')
    model_rbf1.fit(X_train, y_train_2)
    y_pred = model_rbf1.predict(X_test)
    accuracy6 = metrics.accuracy_score(y_true=y_test_2, y_pred=y_pred)
    
    
    results = {'Target Class':letter_uni[i], 'Linear(C=1)':accuracy1, 'Linear(C=5)':accuracy2, 'Linear(C=10)':accuracy3, 'Polynomial(d = 2, C = 5)':accuracy4, 'Polynomial(d = 3, C = 5)':accuracy5, 'RBF(C=5)':accuracy6}
    result.append(results)
    
```
